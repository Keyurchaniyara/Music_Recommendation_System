```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(l)
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key=lambda x:x[1])[1:6]
```

```python
def recommend(music):
    music_index = data1[data1['song_name'] == music].index[0]
    distance = similarity[music_index]
    music_list = sorted(list(enumerate(distance)),reverse = True,key=lambda x:x[1])[1:6]
    
    for i in music_list:
        print(data1.iloc[i[0]].song_name)
        print(data1.iloc[i[0]].spotify_link)
```

```python
recommend('Whole Lotta')
```
